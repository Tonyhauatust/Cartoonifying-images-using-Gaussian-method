```python
import cv2
import easygui 
import os
import tkinter as tk
import tkinter.messagebox
from tkinter import filedialog
from tkinter import *
from PIL import ImageTk, Image

def upload():
    global ImagePath ## Saving the imagepath as a global variable for other functions to retrieve its value
    ImagePath=easygui.fileopenbox() ## Locating the imagepath 


def ReSized6(): ##Function to retrieve the edge of an image
    global ReSized6
    img=cv2.imread(ImagePath,0) ## Reading the image in greycolour
    ReSized6 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2) ## Applying adaptive threshhold method to the image
    ReSized6 = cv2.resize(ReSized6,  (img.shape[0],img.shape[1])) ## Ensuring that the output image has the same size as the original image
    cv2.imshow('Edge retrieved ',ReSized6) 
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
def ReSized5(): ##Function to smooth the image
    global ReSized5
    img=cv2.imread(ImagePath,1) ## Reading the image in colour version
    ReSized5 = cv2.GaussianBlur(img, (5,5),1) ## Applying Gaussian blur method to smoothen the image
    ReSized5 = cv2.resize(ReSized5, (img.shape[0],img.shape[1])) ## Ensuring that the output image has the same size as the original image
    cv2.imshow('Blurred ',ReSized5)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    cv2.destroyAllWindows()


def save_6(ReSized6, ImagePath):
    newName="Edge retrieved image"
    path1 = os.path.dirname(ImagePath) ## Reading the directory name of the original input image
    extension=os.path.splitext(ImagePath)[1] ## Spliting the ext. from the directory name
    path = os.path.join(path1, newName+extension) ## Joining the desired output image name, the directory path and the ext. 
    cv2.imwrite(path, ReSized6) ## Saving the output image
    I = "Image saved by name " + newName +" at "+ path 
    tk.messagebox.showinfo(title=None, message=I) ## Outputing a message box 
    cv2.waitKey(0)
    cv2.destroyAllWindows()

def save_5(ReSized6, ImagePath):
    newName="Blurred image"
    path1 = os.path.dirname(ImagePath)
    extension=os.path.splitext(ImagePath)[1]
    path = os.path.join(path1, newName+extension)
    cv2.imwrite(path, ReSized5)
    I = "Image saved by name " + newName +" at "+ path
    tk.messagebox.showinfo(title=None, message=I)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

top=tk.Tk() ## Defining the root of tkinter
top.geometry('300x700') ## Setting up a window of size 300x700
top.title('Cartoonify Your Image!')
top.configure(background='white') ## Giving a white background
label=Label(top,background='#CDCDCD', font=('calibri',20,'bold'))
uploada=Button(top,text="Retrieving edges of your image",command=lambda:[upload(),ReSized6()],padx=10,pady=5)
## uploada is to define a button to initiate the upload and retriving edge programme
uploadb=Button(top,text="Blurring your image",command=lambda:[upload(),ReSized5()],padx=10,pady=5)
## uploadb is to define a button to initiate the upload and blurring programme 
uploada.configure(background='#364156', foreground='white',font=('calibri',10,'bold'))
## to setup the configuration of the buttons
uploada.pack(side=TOP,pady=50)
uploadb.configure(background='#364156', foreground='white',font=('calibri',10,'bold'))
uploadb.pack(side=TOP,pady=50)
save1=Button(top,text="Save your edge-retrieved image",command=lambda:save_6(ReSized6,ImagePath),padx=30,pady=5)
## save1 to define a button to initiate the 'save the edge-retrieved image' programme
save1.configure(background='#364156', foreground='white',font=('calibri',10,'bold'))
save1.pack(side=TOP,pady=50)
save2=Button(top,text="Save your blurred image",command=lambda:save_5(ReSized6,ImagePath),padx=30,pady=5)
## save2 to define a button to initiate the blurring image programme
save2.configure(background='#364156', foreground='white',font=('calibri',10,'bold'))
save2.pack(side=TOP,pady=50)
destroy_all=Button(top, text="Quit the programme",command=top.destroy,padx=10,pady=5)
## destroy_all to define a button to destroy the window 
destroy_all.configure(background='#364156', foreground='white',font=('calibri',10,'bold'))
destroy_all.pack(side=TOP,pady=50)
top.mainloop()
```
